```python
data_list = filtered_df[['PULocationID', 'DOLocationID']].to_dict(orient='records')
```

```python
from sklearn.feature_extraction import DictVectorizer

vectorizer = DictVectorizer(sparse=False)
feature_matrix = vectorizer.fit_transform(data_list)
feature_matrix
```

```python
filtered_df1 = filtered_df.drop(['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'PULocationID', 'DOLocationID'], axis=1)
filtered_df1.head()
```
